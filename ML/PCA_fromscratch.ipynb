```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
wine = fetch_ucirepo(id=109)

# data (as pandas dataframes)
X = wine.data.features
y = wine.data.targets

df = pd.DataFrame(X)
```

```python
k = 3

covariance_matrix = np.cov(df.T)

eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

sorted_eigenvectors = eigenvectors[:, np.argsort(eigenvalues)[::-1]]

normalized_eigenvectors = sorted_eigenvectors / np.linalg.norm(sorted_eigenvectors)

selected_eigenvectors = sorted_eigenvectors[:,:k]
selected_eigenvectors
pca = np.dot(df - np.mean(df),selected_eigenvectors)

pca
```
